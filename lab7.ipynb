```python
# Ejercicio 1 -
def ordenar_lista_diccionarios(lista_dict, key):
    return sorted(lista_dict, key=lambda x: x[key])

D = [
    {'make': 'Nokia', 'model': 216, 'color': 'Black'},
    {'make': 'Apple', 'model': 2, 'color': 'Silver'},
    {'make': 'Huawei', 'model': 50, 'color': 'Gold'},
    {'make': 'Samsung', 'model': 7, 'color': 'Blue'}
]
print("Lista original:")
for item in D:
    print(item)

print("\nOrdenada por 'model':")
D_ordenado = ordenar_lista_diccionarios(D, 'model')
for item in D_ordenado:
    print(item)

print("\nOrdenada por 'make':")
D_ordenado_make = ordenar_lista_diccionarios(D, 'make')
for item in D_ordenado_make:
    print(item)
```
